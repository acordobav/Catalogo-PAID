# Reducción del Ruido de una Imagen en el Dominio Espacial

Sea $A \in {\rm I\!R}^{m \times n}$ una imagen a escala de grises. (**Nota.** Los valores de los pixeles tienen que estar en el intervalo $[0,1]$, im2double). Se considerará una imagen con ruido aditivo $N \in {\rm I\!R}^{m \times n}$, esta imagen con ruido se representa con la imagen $B \in {\rm I\!R}^{m \times n}$, es decir

\begin{equation}
B = A + N
\end{equation}

```octave
function Z = salt_noise(m, n, p)
  % Funcion para generar un ruido de sal y pimienta
  % m: numero de filas, double
  % n: numero de columnas, double
  % p: probabilidad de cambiar un pixel, double entre 0 y 1
  Z = zeros(m,n);
  p = p*100;
  x = randi([0, 100], [m, n]);
  for i = 1:m
    for j = 1:n
      if (x(i,j) < p)
        Z(i,j) = 1;
      endif
    endfor
  endfor
endfunction
```

```octave
function Z = pepper_noise(m, n, p)
  % Funcion para generar un ruido de sal y pimienta
  % m: numero de filas, double
  % n: numero de columnas, double
  % p: probabilidad de cambiar un pixel, double entre 0 y 1
  Z = zeros(m,n);
  p = p*100;
  x = randi([0, 100], [m, n]);
  for i = 1:m
    for j = 1:n
      if (x(i,j) < p)
        Z(i,j) = -1;
      endif
    endfor
  endfor
endfunction
```

## Técnicas para eliminar ruido aditivo en el dominio espacial
Los métodos para eliminar ruido también se conocen como filtros.

### Filtro del promedio
Sea $W$ un vecindario de un pixel $A_{i,j}$

\begin{equation}
W = \begin{array}{|c|c|c|} \hline
A_{i-1,j-1} & A_{i-1,j} & A_{i-1,j+1} \\\hline
A_{i,j-1}   & A_{i,j}   & A_{i,j+1}  \\\hline
A_{i+1,j-1} & A_{i+1,j} & A_{i+1,j+1} \\\hline
\end{array}
\end{equation}

La matriz $W$ también se le conoce como ventana. Entonces, el filtro promedio está dado por la siguiente fórmula

\begin{equation}
\hat{A}_{i,j} = \dfrac{1}{m \cdot n} \sum_{(r,c) \in W} A_{r,c}
\end{equation}

donde $m \times n$ es el tamaño de $W$, $\hat{A}$ es la imagen restaurada, $r$ y $c$ representan el número de fila y columna, respectivamente.

```octave
function Y = filtro_promedio(A, r)
  % A: imagen original en formato im2double
  % r: double con el radio de la ventana sin contar el centro
  [m,n] = size(A);
  Y = zeros(m,n);
  
  for i = 1:m
    for j = 1:n
      % Calculo de los limites de la ventana actual
      sup = i - r; % Limite superior
      inf = i + r;  % Limite inferior
      izq = j - r; % Limite izquierdo
      der = j + r; % Limite derecho      
      
      % Comprobacion que los limites no sobrepasen 
      % las dimensiones de la imagen original
      if(sup < 1) sup = 1; endif
      if(inf > m) inf = m; endif
      if(izq < 1) izq = 1; endif
      if(der > n) der = n; endif

      % Calculo del promedio
      Y(i,j) = mean(mean(A(sup:inf, izq:der)));
    endfor  
  endfor  
endfunction
```

```octave
% Ejemplo Filtro Promedio
clc; clear; close all;
pkg load image;

A = imread('media/peppers.jpg');
subplot(1,3,1);
imshow(A);
title('Imagen original');

% Se agrega ruido a la imagen original
A = im2double(A);
[m,n] = size(A);
N = 0.1 * randn(m,n);
B = A + N; % Imagen con ruido
subplot(1, 3, 2);
imshow(im2uint8(B));
title('Imagen con ruido');

% radio = 1 -> ventana 3x3
% radio = 2 -> ventana 5x5
% radio = 3 -> ventana 7x7
% ...
% radio = n -> ventana (2n+1)x(2n+1)
C = filtro_promedio(B, 2);
subplot(1, 3, 3);
imshow(im2uint8(C));
title('Imagen filtrada')
```

### Filtro promedio geométrico
$A \in {\rm I\!R}^{p \times q}$ la imagen con ruido. Este filtro está definido por la fórmula

\begin{equation}
\hat{A}_{i,j} = \left[ \prod_{(r,c) \in W} A_{r,c} \right]^{\dfrac{1}{m \cdot n}}
\end{equation}

**Nota.** Este filtro se utiliza frecuentemente cuando se tiene un ruido de tipo Gaussiano.

Se multiplican todos los elementos de cada fila, luego se multiplican estos resultados entre sí, y por último se eleva a la $\dfrac{1}{m\cdot n}$.

```octave
function Y = filtro_promedio_geometrico(A, r)
  % A: imagen original en formato im2double
  % r: double con el radio de la ventana sin contar el centro
  [m,n] = size(A);
  Y = zeros(m,n);
  
  for i = 1:m
    for j = 1:n
      % Calculo de los limites de la ventana actual
      sup = i - r; % Limite superior
      inf = i + r;  % Limite inferior
      izq = j - r; % Limite izquierdo
      der = j + r; % Limite derecho
      
      % Comprobacion que los limites no sobrepasen 
      % las dimensiones de la imagen original
      if(sup < 1) sup = 1; endif
      if(inf > m) inf = m; endif
      if(izq < 1) izq = 1; endif
      if(der > n) der = n; endif

      % Calculo del promedio geometrico
      vecindario = A(sup:inf, izq:der);
      Y(i,j) = prod(prod(vecindario))^(1/numel(vecindario));
    endfor  
  endfor  
endfunction
```

```octave
% Ejemplo del Filtro Promedio Geometrico
clc; clear; close all;
pkg load image;

A = imread('media/peppers.jpg');
subplot(1,3,1);
imshow(A);
title('Imagen original');

% Se agrega ruido a la imagen original
A = im2double(A);
[m,n] = size(A);
N = 0.075 * randn(m,n);
B = A + N; % Imagen con ruido
B = im2uint8(B);
subplot(1, 3, 2);
imshow(B);
title('Imagen con ruido');

% Calculo del Filtro Geometrico
B = im2double(B);
r = 1; % Radio de la ventana, con r = 1 -> ventana = 3x3
C = filtro_promedio_geometrico(B, r);

C = im2uint8(C);
subplot(1,3,3);
imshow(C);
title('Imagen con filtro geometrico');
```

### Filtro armónico promedio
$A \in {\rm I\!R}^{p \times q}$ una imagen con ruido. El filtro armónico promedio está dado por la fórmula

\begin{equation}
\hat{A}_{i,j} = \dfrac{m \cdot n}{\sum_{(r,c) \in W} \dfrac{1}{g(r,c)}}
\end{equation}

**Notas** 
- Si $g(r,c)=0$ entonces se omite.
- Se utiliza este filtro en imágenes con ruido gaussiano y ruido del tipo sal (pixeles blancos).

```octave
function Y = filtro_armonico_promedio(A, r)
  % A: imagen original en formato im2double
  % r: double con el radio de la ventana sin contar el centro
  [m,n] = size(A);
  Y = zeros(m,n);
  
  for i = 1:m
    for j = 1:n
      % Calculo de los limites de la ventana actual
      sup = i - r; % Limite superior
      inf = i + r;  % Limite inferior
      izq = j - r; % Limite izquierdo
      der = j + r; % Limite derecho
      
      % Comprobacion que los limites no sobrepasen 
      % las dimensiones de la imagen original
      if(sup < 1) sup = 1; endif
      if(inf > m) inf = m; endif
      if(izq < 1) izq = 1; endif
      if(der > n) der = n; endif

      vecindario = A(sup:inf, izq:der);
      [m1, n1] = size(vecindario);
      % Calculo del filtro armonico promedio
      sum = 0;
      for x = 1:m1
        for y = 1:n1
          el = vecindario(x,y);
          if (el != 0) sum = sum + 1/el; endif        
        endfor
      endfor
      Y(i,j) = numel(vecindario) / sum;
    endfor  
  endfor
endfunction
```

```octave
% Ejemplo del Filtro Armonico Promedio
clc; clear; close all;
pkg load image;

A = imread('media/peppers.jpg');
subplot(1,3,1);
imshow(A);
title('Imagen original');

% Se agrega ruido a la imagen original
A = im2double(A);
[m,n] = size(A);
N = salt_noise(m,n,0.06);
B = A + N; % Imagen con ruido
B = im2uint8(B);
subplot(1, 3, 2);
imshow(B);
title('Imagen con ruido');

% Calculo del Filtro Armonico Promedio
B = im2double(B);
r = 1; % Radio de la ventana, con r = 1 -> ventana = 3x3
C = filtro_armonico_promedio(B, r);

C = im2uint8(C);
subplot(1,3,3);
imshow(C);
title('Imagen con filtro armonico promedio');
```

### Filtro contraarmónico promedio
$A \in {\rm I\!R}^{p \times q}$ una imagen con ruido, entonces el filtro contraarmónico promedio de orden $R$, se define como

\begin{equation}
\dfrac{\sum_{(r,c) \in W} (A_{r,c})^{R+1}}{\sum_{(r,c) \in W} (A_{r,c})^{R}}
\end{equation}

**Nota.** Este filtro se recomienda para ruidos de tipo sal o pimienta, pero no ambos.

```octave
function Y = filtro_contraarmonico_promedio(A, R, r)
  % A: imagen original en formato im2double
  % R: orden del filtro, formato double
  % r: double con el radio de la ventana sin contar el centro
  [m,n] = size(A);
  Y = zeros(m,n);
  
  for i = 1:m
    for j = 1:n
      % Calculo de los limites de la ventana actual
      sup = i - r; % Limite superior
      inf = i + r;  % Limite inferior
      izq = j - r; % Limite izquierdo
      der = j + r; % Limite derecho
      
      % Comprobacion que los limites no sobrepasen 
      % las dimensiones de la imagen original
      if(sup < 1) sup = 1; endif
      if(inf > m) inf = m; endif
      if(izq < 1) izq = 1; endif
      if(der > n) der = n; endif

      vecindario = A(sup:inf, izq:der);
      
      % Calculo del filtro contraarmonico promedio
      num = sum(sum(vecindario.^(R+1))); % Calculo del numerador
      den = sum(sum(vecindario.^(R))); % Calculo del denominador
      
      if (den != 0) Y(i,j) = num / den; endif
      
    endfor  
  endfor
endfunction
```

```octave
% Ejemplo del Filtro Contraarmonico Promedio
clc; clear; close all;
pkg load image;

A = imread('media/peppers.jpg');
subplot(1,3,1);
imshow(A);
title('Imagen original');

% Se agrega ruido a la imagen original
A = im2double(A);
[m,n] = size(A);
N = pepper_noise(m,n,0.15);
B = A + N; % Imagen con ruido de tipo pimienta
B = im2uint8(B);
subplot(1, 3, 2);
imshow(B);
title('Imagen con ruido');

% Calculo del Filtro Contraarmonico Promedio
B = im2double(B);
R = 2; % Orden del filtro
r = 1; % Radio de la ventana, con r = 1 -> ventana = 3x3
C = filtro_contraarmonico_promedio(B, R, r);

C = im2uint8(C);
subplot(1,3,3);
imshow(C);
title('Filtro contraarmonico promedio');
```

### Filtro de la mediana
Este filtro ordena los valores de los pixeles dentro de un vecindario, encontrando el valor medio, y reemplazando el valor del pixel original con la mediana de este vecindario. 

Ejemplo. Considere el bloque

\begin{array}{|c|c|c|} \hline
9 & 12 & 0 \\\hline
5 & \color{red} 4  & 9 \\\hline
8 & 10 & 7 \\\hline
\end{array}

Se desea cambiar el valor de la entrada central con el valor medio de todo el bloque: <br>
1) Ordenar, de mayor a menor, los valores del bloque
\begin{equation}
0 - 4 - 5 - 7 - \underbrace{8}_{\text{mediana}} - 9 - 9 - 10 - 12
\end{equation}

2) Calcular la mediana

La mediana es $8$.

3) Sustitior el valor central del pixel con la mediana, es decir, 8

\begin{array}{|c|c|c|} \hline
9 & 12 & 0 \\\hline
5 & \color{red} 8  & 9 \\\hline
8 & 10 & 7 \\\hline
\end{array}

**Nota.** Este tipo de filtros se utilizan para limpiar una imagen con ruido "sal y pimienta".

### Filtro del máximo y del mínimo
$A \in {\rm I\!R}^{p \times q}$ una imagen con ruido, entonces

\begin{equation}
\text{1) Máximo: } \hat{A}_{i,j} = max(A_{i,j}/(i,j) \in W) \quad \text{Pimienta}
\end{equation}

\begin{equation}
\text{2) Mínimo: } \hat{A}_{i,j} = min(A_{i,j}/(i,j) \in W) \quad \text{Sal}
\end{equation}

```octave
function Y = filtro_maximo(A, r)
  % A: imagen original en formato im2double
  % r: double con el radio de la ventana sin contar el centro
  [m,n] = size(A);
  Y = zeros(m,n);
  
  for i = 1:m
    for j = 1:n
      % Calculo de los limites de la ventana actual
      sup = i - r; % Limite superior
      inf = i + r;  % Limite inferior
      izq = j - r; % Limite izquierdo
      der = j + r; % Limite derecho
      
      % Comprobacion que los limites no sobrepasen 
      % las dimensiones de la imagen original
      if(sup < 1) sup = 1; endif
      if(inf > m) inf = m; endif
      if(izq < 1) izq = 1; endif
      if(der > n) der = n; endif
      
      % Calculo del filtro maximo
      Y(i,j) = max(max(A(sup:inf, izq:der)));
      
    endfor  
  endfor
endfunction
```

```octave
% Ejemplo del Filtro Maximo
clc; clear; close all;
pkg load image;

A = imread('media/peppers.jpg');
subplot(1,3,1);
imshow(A);
title('Imagen original');

% Se agrega ruido a la imagen original
A = im2double(A);
[m,n] = size(A);
N = pepper_noise(m,n,0.2);
B = A + N; % Imagen con ruido de tipo pimienta
B = im2uint8(B);
subplot(1, 3, 2);
imshow(B);
title('Imagen con ruido');

% Calculo del Filtro Maximo
B = im2double(B);
r = 1; % Radio de la ventana, con r = 1 -> ventana = 3x3
C = filtro_maximo(B, r);

C = im2uint8(C);
subplot(1,3,3);
imshow(C);
title('Filtro maximo');
```

```octave
function Y = filtro_minimo(A, r)
  % A: imagen original en formato im2double
  % r: double con el radio de la ventana sin contar el centro
  [m,n] = size(A);
  Y = zeros(m,n);
  
  for i = 1:m
    for j = 1:n
      % Calculo de los limites de la ventana actual
      sup = i - r; % Limite superior
      inf = i + r;  % Limite inferior
      izq = j - r; % Limite izquierdo
      der = j + r; % Limite derecho
      
      % Comprobacion que los limites no sobrepasen 
      % las dimensiones de la imagen original
      if(sup < 1) sup = 1; endif
      if(inf > m) inf = m; endif
      if(izq < 1) izq = 1; endif
      if(der > n) der = n; endif
      
      % Calculo del filtro minimo
      Y(i,j) = min(min(A(sup:inf, izq:der)));
      
    endfor  
  endfor
endfunction
```

### Filtro del punto medio

\begin{equation}
A_{i,j} = \dfrac{max(A_{i,j}/(i,j) \in W) + min(A_{i,j}/(i,j) \in W)}{2}
\end{equation}

```octave
function Y = filtro_punto_medio(A, r)
  % A: imagen original en formato im2double
  % r: double con el radio de la ventana sin contar el centro
  [m,n] = size(A);
  Y = zeros(m,n);
  
  for i = 1:m
    for j = 1:n
      % Calculo de los limites de la ventana actual
      sup = i - r; % Limite superior
      inf = i + r;  % Limite inferior
      izq = j - r; % Limite izquierdo
      der = j + r; % Limite derecho
      
      % Comprobacion que los limites no sobrepasen 
      % las dimensiones de la imagen original
      if(sup < 1) sup = 1; endif
      if(inf > m) inf = m; endif
      if(izq < 1) izq = 1; endif
      if(der > n) der = n; endif
      
      % Calculo del filtro minimo
      vecindario = A(sup:inf, izq:der);
      Y(i,j)=(min(min(vecindario))+max(max(vecindario)))/2;
      
    endfor  
  endfor
endfunction
```

```octave
% Ejemplo del Filtro Punto Medio
clc; clear; close all;
pkg load image;

A = imread('media/peppers.jpg');
subplot(1,3,1);
imshow(A);
title('Imagen original');

% Se agrega ruido a la imagen original
A = im2double(A);
[m,n] = size(A);
N = 0.3*rand(m,n);
B = A + N; % Imagen con ruido de tipo sal y pimienta
B = im2uint8(B);
subplot(1, 3, 2);
imshow(B);
title('Imagen con ruido');

% Calculo del Filtro Punto Medio
B = im2double(B);
r = 1; % Radio de la ventana, con r = 1 -> ventana = 3x3
C = filtro_punto_medio(B, r);

C = im2uint8(C);
subplot(1,3,3);
imshow(C);
title('Filtro punto medio');
```
