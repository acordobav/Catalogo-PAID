# Introducci√≥n a GNU Octave

```octave
% Cargar una imagen
X = imread('media/lena.jpg');

% Mostrar imagen
imshow(X);
```

```octave
% Tama√±o de la imagen
[m ,n, r] = size(X)
```

```octave
% Informaci√≥n de la imagen
imfinfo('media/lena.jpg');
```

```octave
% Mostrar las primeras cinco filas y cinco columnas del canal rojo
R5 = X(1:5, 1:5, 1);
```

El primer argumento corresponde al n√∫mero de filas, el segundo al n√∫mero de columnas, y el tercero al n√∫mero del canal. Los canales de este tipo de im√°genes vienen en este orden respectivo: RED (1), GREEN (2), BLUE (3).

```octave
% Convertir de formato imagen a formato double
R5_2 = im2double(R5);

% Guardar los canales en matrices aparte
R = X(:, :, 1); % Canal rojo
G = X(:, :, 2); % Canal verde
B = X(:, :, 3); % Canal azul

imshow(R);
```

Se muestra en blanco y negro ya que se identifica como una imagen de un solo canal, por lo que se muestra la imagen en una escala de grises.

```octave
% Guardar imagen con solo el canal rojo habilitado y los dem√°s en cero
Y = zeros(m, n, 3);
Y(:, :, 1) = R; % Guardar la informacion del canal rojo
Y = uint8(Y); % Se convierte a un formato de 8 bits

imshow(Y);
```

```octave
% Guardar la imagen
imwrite(Y, 'media/canal_rojo.jpg', 'quality', 100);
```
