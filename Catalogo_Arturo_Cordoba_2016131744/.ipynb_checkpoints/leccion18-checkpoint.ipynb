Usando el domimio espacial, además de asumir que la imagen **no** tuene ruido aditivo, entonces

\begin{equation}
g = f * h
\end{equation}

donde $f$: imagen original, $h$: máscara que representa un filtro paso bajo, $g$: imagen difuminada.

Usando el dominio de la frecuencia

\begin{equation}
G = F \odot H
\end{equation}

donde $G$, $F$, $H$ son las DFT-2D de $g$, $f$, $h$ y $\odot$ es la multiplicación puntual de matrices.


Para una entrada en específico, tenemos

\begin{equation}
G(u,v) = F(u,v) \cdot H(u,v)
\end{equation}

Si $H(u,v) \neq 0$, entonces

\begin{equation}
F(u,v)=\dfrac{G(u,v)}{H(u,v)}
\end{equation}

**Nota.** Usando el filtro paso bajo, se obtiene que $H$ es

![title](media/nota-difuminacion.svg)


Una manera de eliminar una difuminación de una imagen es usando la siguiente fórmula: Sea $g$ la imagen difuminada y sea $h$ la máscara a utilizar. Sean $G$ y $H$ la DFT-2D de $g$ y $h$. Entonces

\begin{equation}
1) \quad \hat{F}(u,v)=\left\lbrace
\begin{array}{lcr}
     \dfrac{G(u,v)}{H(u,v)} \quad \text{si } |H(u,v)| \geq \text{tol} \\
     G(u,v) \quad \text{si } |H(u,v)| < \text{tol} 
\end{array}
\right.
\end{equation}

**Nota.** No utilizar un filtro paso bajo ideal para reconstruir una imagen difuminada, es mejor utilizar gaussiano o butterworth.

## Filtro de Wiener
Sea $G$ y $H$ la DFT-2D de $g$ (imagen difuminada) y $h$ (filtro paso bajo). Entonces

\begin{equation}
2) \quad \hat{F}(u,v) = \left[ \dfrac{1}{H(u,v)} \cdot \dfrac{|H(u,v)|^2}{|H(u,v)|^2+k} \right] G(u,v)
\end{equation}

donde $k$ es una constante que aproxima la cantidad de ruido.

# Problema 1
Dada la imagen del camarógrafo difuminado, trate de eliminar la difuminación utilizando los filtros paso bajo Gaussiano y Butterworth, usando la fórmula **1**.

```octave
function H = fpb_gaussiano(m,n,s)
  % Funcion para generar un filtro paso bajo gaussiano
  % m: numero de filas, double
  % n: numero de columnas, double
  % s: parametro de corte, debe ser un double > 0
  % H: imagen en formato double con el filtro paso bajo gaussiano
  dist = zeros(m,n);
  m1 = floor(m/2); n1 = floor(n/2);
  for i = 1:m
    for j = 1:n
      dist(i,j) = sqrt((i-m1)^2 + (j-n1)^2);
    endfor
  endfor
  H = exp(-(dist.^2)/(2*s^2));
endfunction
```

```octave
function H = fpb_butterworth(m,n,d0,o)
  % Funcion para generar un filtro paso bajo butterworth
  % m: numero de filas, double
  % n: numero de columnas, double
  % d0: frecuencia de corte, double
  % o: orden del filtro butterworth, double 
  % H: imagen en formato double con el filtro paso bajo butterworth
  dist = zeros(m, n);
  m1 = floor(m/2); n1 = floor(n/2);
  for i = 1:m
    for j = 1:n
      dist(i, j) = sqrt((i-m1)^2 + (j-n1)^2);
    endfor
  endfor
  H = 1./(1+(d0./dist).^(-2*o));
endfunction
```

```octave
function F = filtro_difuminado(g,H,tol)
  % Filtro para imagenes difuminadas utilizando un filtro paso bajo gaussiano
  % g: imagen difuminada en formato im2double
  % H: filtro gaussiano o butterworth en formato im2double
  % tol: double > 0, tolerancia para el algoritmo de filtrado 
  % F: imagen filtrada en formato im2double
  g = im2double(g);
  [m,n] = size(g);
  G = fftshift(fft2(g)); % Transformada DFT-2D de g
  
  % Filtrado de la imagen difuminada
  F = zeros(m,n);
  for u = 1:m
    for v = 1:n
      if (abs(H(u,v)) >= tol)
        F(u,v) =  G(u,v)/H(u,v);
      else   
        F(u,v) = G(u,v);
      endif
    endfor
  endfor

  % Transformacion al dominio espacial
  F = fftshift(F);
  F = abs(ifft2(F));
endfunction
```

```octave
clc; clear; close all;
x = imread('media/camarografo_dif.jpg');
x = im2double(x);
[m,n] = size(x);

tol = 0.05;

s = 12;
H1 = fpb_gaussiano(m,n,s); % Filtro gaussiano paso bajo
F1 = filtro_difuminado(x,H1,tol);

d0 = 18;
o = 2;
H2 = fpb_butterworth(m,n,d0,o);
F2 = filtro_difuminado(x,H2,tol);

subplot(1,3,1);
imshow(x);
title('Imagen Original')

subplot(1,3,2);
imshow(F1);
title('Imagen filtrada - Gaussiano');

subplot(1,3,3);
imshow(F2);
title('Imagen filtrada - Butterworth');
```

```octave
X = fftshift(fft2(x)); % Transformada DFT-2D de x
subplot(1,3,1);
imshow(log(1+abs(X)), []);
title('Frecuencia imagen difuminada');

s = 12;
H1 = fpb_gaussiano(m,n,s); % Filtro gaussiano paso bajo

subplot(1,3,2);
imshow(H1);
title('Filtro gaussiano paso bajo');

subplot(1,3,3);
imshow(H2);
title('Filtro butterworth paso bajo');
```

# Problema 2
Repetir el experimento del problema 1, usando la fórmula **2**.

```octave
function F = filtro_wiener(g,H,k)
  % Filtro para imagenes difuminadas utilizando un filtro paso bajo gaussiano
  % g: imagen difuminada en formato im2double
  % H: filtro gaussiano o butterworth en formato im2double
  % k: double que aproxima la cantidad de ruido
  % F: imagen filtrada en formato im2double
  g = im2double(g);
  [m,n] = size(g);
  G = fftshift(fft2(g)); % Transformada DFT-2D de g
  
  % Filtrado de la imagen difuminada
  %F = ((1./H).*((abs(H).^2)./(abs(H).^2+k))).*(abs(G).^2);
  
  F = zeros(m,n);
  for u = 1:m
    for v = 1:n
      H_uv = H(u,v);
      G_uv = G(u,v);
      F(u,v) = ((1/H_uv)*((abs(H_uv)^2)/((abs(H_uv)^2)+k)))*(G_uv);
    endfor
  endfor
  
  % Transformacion al dominio espacial
  F = fftshift(F);
  F = abs(ifft2(F));
endfunction
```

```octave
clc; clear; close all;
x = imread('media/camarografo_dif.jpg');
x = im2double(x);
[m,n] = size(x);

X = fftshift(fft2(x)); % Transformada DFT-2D de x
subplot(1,2,1);
imshow(log(1+abs(X)), []);
title('Frecuencia imagen difuminada');

s = 12;
H1 = fpb_gaussiano(m,n,s); % Filtro gaussiano paso bajo

subplot(1,2,2);
imshow(H1);
title('Filtro gaussiano paso bajo');
```

```octave
k1 = 0.1;
F1 = filtro_wiener(x,H1,k1);

k2 = 0.3;
F2 = filtro_wiener(x,H1,k2);

k3 = 0.5;
F3 = filtro_wiener(x,H1,k3);

subplot(2,2,1);
imshow(x);
title('Imagen difuminada');

subplot(2,2,2);
imshow(F1);
title(['Wiener Gaussiano k = ' num2str(k1)]);

subplot(2,2,3);
imshow(F2);
title(['Wiener Gaussiano k = ' num2str(k2)]);

subplot(2,2,4);
imshow(F3);
title(['Wiener Gaussiano k = ' num2str(k3)]);
```

```octave
subplot(1,2,1);
imshow(log(1+abs(X)), []);
title('Frecuencia imagen difuminada');

d0 = 18;
o = 3;
H2 = fpb_butterworth(m,n,d0,o); % Filtro butterworth paso bajo

subplot(1,2,2);
imshow(H2);
title('Filtro butterworth paso bajo');
```

```octave
k1 = 0.1;
F1 = filtro_wiener(x,H2,k1);

k2 = 0.3;
F2 = filtro_wiener(x,H2,k2);

k3 = 0.5;
F3 = filtro_wiener(x,H2,k3);

subplot(2,2,1);
imshow(x);
title('Imagen difuminada');

subplot(2,2,2);
imshow(F1);
title(['Wiener Butterworth k = ' num2str(k1)]);

subplot(2,2,3);
imshow(F2);
title(['Wiener Butterworth k = ' num2str(k2)]);

subplot(2,2,4);
imshow(F3);
title(['Wiener Butterworth k = ' num2str(k3)]);
```
